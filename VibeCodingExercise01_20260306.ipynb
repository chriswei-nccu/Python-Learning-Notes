```python
import os
import pandas as pd
from google.colab import files
from IPython.display import display, Markdown

# 顯示標題
display(Markdown("# Vibe Coding 檔案匯整"))

print("請選取多個 Excel 檔案（支援 .xlsx、.xls）")
uploaded = files.upload()

# 檢查是否有上傳檔案
if not uploaded:
    print("沒有上傳任何檔案，程式結束。")
else:
    all_data = []

    for file_name in uploaded.keys():
        try:
            print(f"讀取中：{file_name}")

            # 讀取 Excel 檔案的第一個工作表
            df = pd.read_excel(file_name)

            # 增加來源檔名欄位，方便之後辨識資料來自哪個檔案
            df["來源檔名"] = file_name

            all_data.append(df)

        except Exception as e:
            print(f"檔案 {file_name} 讀取失敗：{e}")

    # 檢查是否有成功讀到資料
    if not all_data:
        print("所有檔案都無法讀取，程式結束。")
    else:
        # 合併所有資料
        merged_df = pd.concat(all_data, ignore_index=True)

        print("\n合併完成！")
        print(f"總筆數：{len(merged_df)}")
        print(f"總欄位數：{len(merged_df.columns)}")

        # 顯示前 5 筆
        print("\n合併結果（前 5 筆）：")
        display(merged_df.head(5))

        # 輸出成新的 Excel 檔案
        output_file = "merged_result.xlsx"
        merged_df.to_excel(output_file, index=False)

        print(f"\n已產生合併檔案：{output_file}")

        # 自動下載
        files.download(output_file)
```

```python
html_content = """
<!DOCTYPE html>
<html>
<head>
    <title>Vibe Coding 檔案匯整報告</title>
    <style>
        body {{ font-family: Arial, sans-serif; margin: 20px; }}
        h1 {{ color: #333; }}
        pre {{ background-color: #eee; padding: 10px; border-radius: 5px; overflow-x: auto; }}
        table {{ width: 100%; border-collapse: collapse; margin-top: 20px; }}
        th, td {{ border: 1px solid #ddd; padding: 8px; text-align: left; }}
        th {{ background-color: #f2f2f2; }}
        .download-link {{ margin-top: 20px; display: block; }}
    </style>
</head>
<body>
    <h1>Vibe Coding 檔案匯整報告</h1>

    <h2>合併程式碼</h2>
    <pre>
import os
import pandas as pd
from google.colab import files
from IPython.display import display, Markdown

# 顯示標題
display(Markdown("# Vibe Coding 檔案匯整"))

print("請選取多個 Excel 檔案（支援 .xlsx、.xls）")
uploaded = files.upload()

# 檢查是否有上傳檔案
if not uploaded:
    print("沒有上傳任何檔案，程式結束。")
else:
    all_data = []

    for file_name in uploaded.keys():
        try:
            print(f"讀取中：{{file_name}}")

            # 讀取 Excel 檔案的第一個工作表
            df = pd.read_excel(file_name)

            # 增加來源檔名欄位，方便之後辨識資料來自哪個檔案
            df["來源檔名"] = file_name

            all_data.append(df)

        except Exception as e:
            print(f"檔案 {{file_name}} 讀取失敗：{{e}}")

    # 檢查是否有成功讀到資料
    if not all_data:
        print("所有檔案都無法讀取，程式結束。")
    else:
        # 合併所有資料
        merged_df = pd.concat(all_data, ignore_index=True)

        print("\n合併完成！")
        print(f"總筆數：{{len(merged_df)}}")
        print(f"總欄位數：{{len(merged_df.columns)}}")

        # 顯示前 5 筆
        print("\n合併結果（前 5 筆）：")
        display(merged_df.head(5))

        # 輸出成新的 Excel 檔案
        output_file = "merged_result.xlsx"
        merged_df.to_excel(output_file, index=False)

        print(f"\n已產生合併檔案：{{output_file}}")

        # 自動下載
        files.download(output_file)
    </pre>

    <h2>合併結果概覽 (前 5 筆)</h2>
    <p>總筆數：{total_rows}</p>
    <p>總欄位數：{total_columns}</p>
    {df_html}

    <h2>已產生合併檔案</h2>
    <p>檔案名稱：{output_file_name}</p>
    <p>您也可以點擊這裡下載合併後的 Excel 檔案：
        <a href="{output_file_name}" download class="download-link">下載 {output_file_name}</a>
    </p>
</body>
</html>
"""

# Prepare data for HTML
total_rows = len(merged_df)
total_columns = len(merged_df.columns)
df_html = merged_df.head(5).to_html(index=False)
output_file_name = 'merged_result.xlsx'

# Format the HTML content
final_html = html_content.format(
    total_rows=total_rows,
    total_columns=total_columns,
    df_html=df_html,
    output_file_name=output_file_name
)

# Save to an HTML file
report_html_file = "merged_report.html"
with open(report_html_file, "w", encoding="utf-8") as f:
    f.write(final_html)

print(f"已產生 HTML 報告：{report_html_file}")

# Auto-download the HTML report
files.download(report_html_file)
```
